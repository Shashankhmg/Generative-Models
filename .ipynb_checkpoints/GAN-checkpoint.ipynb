```python
def preprocess_images(images):
#normlaizing the images
  images = images.reshape((images.shape[0], 28, 28, 1)) / 255.
  return np.where(images > .5, 1.0, 0.0).astype('float32')

train_images = preprocess_images(train_images)
test_images = preprocess_images(test_images)
train_images.shape
```

```python
# Function to plot images
def plot_images(images, num_images):
    plt.figure(figsize=(10, 10))
    for i in range(num_images):
        plt.subplot(5, 5, i + 1)
        plt.imshow(images[i].reshape(28, 28), cmap='gray')
        plt.axis('off')
    plt.show()

# Plot some of the training images
plot_images(train_images, 10)
```

```python
train_size = 60000
batch_size = 32
test_size = 10000
```

```python
train_dataset = (tf.data.Dataset.from_tensor_slices(train_images)
                 .shuffle(train_size).batch(batch_size))
test_dataset = (tf.data.Dataset.from_tensor_slices(test_images)
                .shuffle(test_size).batch(batch_size))
```

**Creating a GAN Model**

```python
from tensorflow.keras.layers import Dense, Reshape
from tensorflow.keras.models import Sequential

def generator(latent_dim, n_outputs=(28, 28, 1)):
    model = Sequential(name="Generator")  # Model

    # Input Layer: The latent vector
    model.add(Dense(7*7*256, activation='relu', kernel_initializer='he_uniform', input_shape=(latent_dim,), name='Generator-Hidden-Layer-1'))  # Hidden Layer 1
    model.add(Dense(7*7*128, activation='relu', kernel_initializer='he_uniform', name='Generator-Hidden-Layer-2'))  # Hidden Layer 2
    model.add(Dense(14*14*64, activation='relu', kernel_initializer='he_uniform', name='Generator-Hidden-Layer-3'))  # Hidden Layer 3
    model.add(Dense(28*28*1, activation='tanh', kernel_initializer='glorot_uniform', name='Generator-Output-Layer'))  # Output Layer
    # Reshape the final output to (28, 28, 1)
    model.add(Reshape(n_outputs, name='Generator-Reshape-Layer'))


    return model

# Instantiate
latent_dim = 100
gen_model = generator(latent_dim)

# Print the model summary
gen_model.summary()
```

**Set up the discriminator model**



```python
import tensorflow as tf
from tensorflow.keras import layers, Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import plot_model

def discriminator(input_shape=(28, 28, 1)):
    model = Sequential(name="Discriminator")  # Model

    # Add Layers
    model.add(Flatten(input_shape=input_shape, name='Discriminator-Input-Flatten'))  # Flatten the input
    model.add(Dense(32, activation='relu', kernel_initializer='he_uniform', name='Discriminator-Hidden-Layer-1'))  # Hidden Layer
    model.add(Dense(16, activation='relu', kernel_initializer='he_uniform', name='Discriminator-Hidden-Layer-2'))  # Hidden Layer
    model.add(Dense(1, activation='sigmoid', name='Discriminator-Output-Layer'))  # Output Layer

    # Compile the model
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

# Instantiate
dis_model = discriminator()

# Show model summary and plot model diagram
dis_model.summary()
#plot_model(dis_model, show_shapes=True, show_layer_names=True, dpi=400)
```

```python
def def_gan(generator, discriminator):

    # We don't want to train the weights of discriminator at this stage. Hence, make it not trainable
    discriminator.trainable = False

    # Combine
    model = Sequential(name="GAN") # GAN Model
    model.add(generator) # Add Generator
    model.add(discriminator) # Add Disriminator

    # Compile the model
    model.compile(loss='binary_crossentropy', optimizer='adam')
    return model

# Instantiate
gan_model = def_gan(gen_model, dis_model)

# Show model summary and plot model diagram
gan_model.summary()
#plot_model(gan_model, show_shapes=True, show_layer_names=True, dpi=400)
```

**Preparing inputs for the generator and the discriminator**


```python
def latent_points(latent_dim, n):
    """
    Generate random noise vectors of given dimension and number from a standard normal distribution.

    Parameters:
    latent_dim (int): Dimension of the latent space.
    n (int): Number of noise vectors to generate.

    Returns:
    noise (numpy array): Array of generated noise vectors.
    """
    noise = np.random.normal(0, 1, size=(n, latent_dim, ))
    return noise
fake = latent_points(100,1)
print(fake.shape)
```

```python
# Use the generator to generate n fake examples together with class labels
def fake_samples(generator, n, latent_dim = 100):

    #print(n)

    # Generate points in latent space
    latent_output = latent_points(latent_dim, n)
    latent_output_tf = tf.convert_to_tensor(latent_output, dtype=tf.float32)  # Convert to TensorFlow tensor
    # Predict outputs (i.e., generate fake samples)
    X = generator.predict(latent_output_tf)
    print(X.shape)

    # Create class labels
    y = np.zeros((n, 1))
    return X, y
```

**Model training and evaluation**


```python
import matplotlib.pyplot as plt

def performance_summary(epoch, generator, discriminator, latent_dim, n=100):

    # Initialize accumulators for real and fake accuracies
    real_accuracy_accumulator = []
    fake_accuracy_accumulator = []

    # Evaluate the discriminator on real data
    for images_batch in train_dataset:
        y_real = np.ones((images_batch.shape[0], 1))  # Real labels
        real_loss, real_accuracy = discriminator.evaluate(images_batch, y_real, verbose=0)
        real_accuracy_accumulator.append(real_accuracy)

    real_accuracy = np.mean(real_accuracy_accumulator)  # Compute average real accuracy
    # Get fake (generated) samples
    x_fake, y_fake = fake_samples(generator, n, latent_dim)
    fake_loss, fake_accuracy = discriminator.evaluate(x_fake, y_fake, verbose=0)

    # Summarize discriminator performance
    print(f"Epoch number: {epoch}")
    print(f"Discriminator Accuracy on REAL points: {real_accuracy}")
    print(f"Discriminator Accuracy on FAKE (generated) points: {fake_accuracy}")

    # Plot some generated images
    plt.figure(figsize=(10, 10))
    for i in range(25):
        plt.subplot(5, 5, i + 1)
        plt.imshow(x_fake[i, :, :, 0], cmap='gray')
        plt.axis('off')
    plt.suptitle(f'Generated Images at Epoch {epoch}', fontsize=20)
    plt.savefig(f'real_images_epoch_{epoch}.png')
    plt.show()


    # Plot some real images
    plt.figure(figsize=(10, 10))
    for i, real_image_batch in enumerate(train_dataset.take(1)):  # Take one batch of real images
        for j in range(25):
            plt.subplot(5, 5, j + 1)
            plt.imshow(real_image_batch[j, :, :, 0], cmap='gray')
            plt.axis('off')
    plt.suptitle(f'Real Images at Epoch {epoch}', fontsize=20)
    plt.savefig(f'generated_images_epoch_{epoch}.png')
    plt.show()



```

**Finally, the training function**



```python
x_fake, y_fake = fake_samples(gen_model, 128, latent_dim = 100)
```

```python
def train(g_model, d_model, gan_model, latent_dim = 100, n_epochs=30, n_batch=32, n_eval=1000):



    # We will manually enumare epochs
    for i in range(n_epochs):

        # Discriminator training
        # Prep real samples
        x_real= train_dataset
        # Prep fake (generated) samples
        x_fake, y_fake = fake_samples(g_model, n_batch, latent_dim)
        print(f"Generator generated fake random noise for epoch {i}")
        #print(x_fake.shape)
        #print(y_fake.shape)
        # Train the discriminator using real and fake samples
        for image_batch in x_real:
          y_real = np.ones((image_batch.shape[0], 1))  # NumPy ones for real samples
          d_model.train_on_batch(image_batch, y_real)
        print(f"Discriminator trained on real data for epoch {i}")
        d_model.train_on_batch(x_fake, y_fake)
        print(f"Discriminator trained on fake datafor epoch {i}")

        # Generator training
        # Get points from the latent space to be used as inputs for the generator
        x_gan = latent_points(latent_dim, n = 32)
        print(f"random points generated from the latent space for epoch {i}")

        #print(x_gan.shape)
        # While we are generating fake samples,
        # we want GAN generator model to create examples that resemble the real ones,
        # hence we want to pass labels corresponding to real samples, i.e. y=1, not 0.
        y_gan = np.ones((n_batch, 1))
        #print(y_gan.shape)

        # Train the generator via a composite GAN model
        gan_model.train_on_batch(x_gan, y_gan)
        print(f"GAN trained on real data for epoch {i}")
        performance_summary(i, g_model, d_model, latent_dim)

```

**Results**

```python
# Train GAN model
train(gen_model, dis_model, gan_model)
```
