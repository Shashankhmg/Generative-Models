```python
import tensorflow as tf
from tensorflow.keras import layers, Model
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
```

```python
##Load and Preprocess the Data
```

```python
# Load the dataset
(train_images, _), (test_images, _) = mnist.load_data()

# Normalize the images to [0, 1] range
train_images = train_images.astype('float32') / 255.0
test_images = test_images.astype('float32') / 255.0

# Add a channel dimension
train_images = np.expand_dims(train_images, axis=-1)
test_images = np.expand_dims(test_images, axis=-1)

# Create TensorFlow datasets
train_dataset = tf.data.Dataset.from_tensor_slices(train_images).shuffle(60000).batch(128)
test_dataset = tf.data.Dataset.from_tensor_slices((test_images, test_labels)).batch(128)
```

```python
##Define the VAE Model
```

```python
latent_dim = 2  # Dimensionality of the latent space

# Encoder
class Encoder(Model):
    def __init__(self, latent_dim):
        super(Encoder, self).__init__()
        self.conv1 = layers.Conv2D(32, (3, 3), activation='relu', strides=2, padding='same')
        self.conv2 = layers.Conv2D(64, (3, 3), activation='relu', strides=2, padding='same')
        self.flatten = layers.Flatten()
        self.dense = layers.Dense(16, activation='relu')
        self.mean = layers.Dense(latent_dim)
        self.log_var = layers.Dense(latent_dim)

    def call(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.flatten(x)
        x = self.dense(x)
        mean = self.mean(x)
        log_var = self.log_var(x)
        return mean, log_var

# Decoder
class Decoder(Model):
    def __init__(self, latent_dim):
        super(Decoder, self).__init__()
        self.dense1 = layers.Dense(7 * 7 * 64, activation='relu')
        self.reshape = layers.Reshape((7, 7, 64))
        self.conv_trans1 = layers.Conv2DTranspose(64, (3, 3), activation='relu', strides=2, padding='same')
        self.conv_trans2 = layers.Conv2DTranspose(32, (3, 3), activation='relu', strides=2, padding='same')
        self.conv_trans3 = layers.Conv2DTranspose(1, (3, 3), activation='sigmoid', padding='same')

    def call(self, x):
        x = self.dense1(x)
        x = self.reshape(x)
        x = self.conv_trans1(x)
        x = self.conv_trans2(x)
        x = self.conv_trans3(x)
        return x

# VAE Model
class VAE(Model):
    def __init__(self, encoder, decoder):
        super(VAE, self).__init__()
        self.encoder = encoder
        self.decoder = decoder

    def call(self, x):
        mean, log_var = self.encoder(x)
        epsilon = tf.random.normal(shape=tf.shape(mean))
        z = mean + tf.exp(0.5 * log_var) * epsilon
        reconstructed = self.decoder(z)
        return reconstructed

# Instantiate the encoder, decoder, and VAE
encoder = Encoder(latent_dim)
decoder = Decoder(latent_dim)
vae = VAE(encoder, decoder)
```

```python
##Define the Loss Function and Optimizer
```

```python
optimizer = tf.keras.optimizers.Adam()

def compute_loss(model, x):
    mean, log_var = model.encoder(x)
    epsilon = tf.random.normal(shape=tf.shape(mean))
    z = mean + tf.exp(0.5 * log_var) * epsilon
    x_logit = model.decoder(z)

    #for the decoder
    reconstruction_loss = tf.reduce_mean(
        tf.keras.losses.binary_crossentropy(x, x_logit)
    )
    reconstruction_loss *= 28 * 28
    
    kl_loss = 1 + log_var - tf.square(mean) - tf.exp(log_var)
    kl_loss = tf.reduce_mean(kl_loss) * -0.5

    total_loss = reconstruction_loss + kl_loss
    return total_loss

@tf.function
def train_step(model, x, optimizer):
    with tf.GradientTape() as tape:
        loss = compute_loss(model, x)
    gradients = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))
    return loss
```

```python
##Training loop
```

```python
import time

# Training loop with timer and latent space recording
epochs = 30
latent_spaces = []

for epoch in range(epochs):
    start_time = time.time()
    
    for train_x in train_dataset:
        train_step(vae, train_x, optimizer)
    
    # Record the latent space at the end of each epoch
    epoch_latents = []
    epoch_labels = []
    for test_x, test_y in test_dataset:
        mean, _ = vae.encoder(test_x)
        epoch_latents.append(mean)
        epoch_labels.append(test_y)
    
    epoch_latents = np.concatenate(epoch_latents, axis=0)
    epoch_labels = np.concatenate(epoch_labels, axis=0)
    latent_spaces.append((epoch_latents, epoch_labels))
    
    loss = tf.reduce_mean([compute_loss(vae, test_x) for test_x, _ in test_dataset])
    
    end_time = time.time()
    epoch_time = end_time - start_time
    
    print(f'Epoch {epoch + 1}, Loss: {loss.numpy()}, Time: {epoch_time:.2f} seconds')

```

```python
##Visualize the Latent Space
```

```python
def plot_latent_space(vae, n=30, figsize=15):
    digit_size = 28
    scale = 2.0
    figure = np.zeros((digit_size * n, digit_size * n))
    grid_x = np.linspace(-scale, scale, n)
    grid_y = np.linspace(-scale, scale, n)

    for i, yi in enumerate(grid_x):
        for j, xi in enumerate(grid_y):
            z_sample = np.array([[xi, yi]])
            x_decoded = vae.decoder(z_sample)
            digit = tf.reshape(x_decoded[0], (digit_size, digit_size)).numpy()
            figure[i * digit_size: (i + 1) * digit_size,
                   j * digit_size: (j + 1) * digit_size] = digit

    plt.figure(figsize=(figsize, figsize))
    plt.imshow(figure, cmap='Greys_r')
    plt.axis('off')
    plt.show()

# Plot the learned latent space
plot_latent_space(vae)
```
